Lengkapkan identiti robot dan gunakan `input()` untuk mendapatkan nama jurutera.

````python
# ==========================================================
# TASK 1: IDENTITI ROBOT
# ==========================================================

robot_name = "FIX_ME"      # Contoh: "KV-TRON Rondaan Keselamatan"
robot_id = None            # Contoh: 101
operation_mode = "FIX_ME"  # Contoh: "Pemeriksaan Bengkel"
robot_location = "FIX_ME"  # Contoh: "Bengkel TVET A"

# FIX ME: Tukar kepada input("Masukkan nama jurutera: ")
engineer_name = "FIX_ME"

print("\nIDENTITI ROBOT")
print("Nama Robot  :", robot_name)
print("ID Robot    :", robot_id)
print("Mod Operasi :", operation_mode)
print("Lokasi      :", robot_location)
print("Jurutera    :", engineer_name)

identity_ready = (
    robot_name != "FIX_ME" and robot_id is not None
    and operation_mode != "FIX_ME" and robot_location != "FIX_ME"
    and engineer_name != "FIX_ME"
)
print("[LULUS] Identiti lengkap." if identity_ready else "[BELUM LENGKAP] Baiki TASK 1.")
````

# TASK 2: Keputusan Keselamatan

Lengkapkan function supaya menghasilkan output berikut:

| Keadaan | Output Wajib |
|---|---|
| Bateri `< 20%` | `KEMBALI KE PENGECAS` |
| Jarak `< 10 cm` | `HENTI KECEMASAN` |
| Jarak `<= 30 cm` | `AMARAN: KURANGKAN KELAJUAN` |
| Selainnya | `MISI DILULUSKAN` |

````python
# ==========================================================
# TASK 2: FUNCTION KESELAMATAN
# ==========================================================

def decide_mission(battery_level, distance):
    # FIX ME: Lengkapkan threshold dan mesej output
    if battery_level < 0:
        return "FIX_ME"
    elif distance < 0:
        return "FIX_ME"
    elif distance <= 0:
        return "FIX_ME"
    else:
        return "FIX_ME"
````

# TASK 3: Rondaan Autonomi dan Rekod Suhu

Rondaan hanya berlaku jika output safety ialah `MISI DILULUSKAN`. Setiap zon menggunakan `15%` bateri.

````python
# ==========================================================
# TASK 3: FUNCTION RONDAAN
# ==========================================================

zones = ["Zon Pintu Masuk", "Zon Kimpalan", "Zon Simpanan Alatan", "Zon Operasi Mesin"]
zone_readings = {
    "Zon Pintu Masuk": 29,
    "Zon Kimpalan": 43,
    "Zon Simpanan Alatan": 30,
    "Zon Operasi Mesin": 38
}

def run_patrol(starting_battery, decision):
    inspection_log = []
    warning_zones = []
    battery_level = starting_battery

    # FIX ME: Rondaan hanya bermula jika decision betul
    if decision != "FIX_ME":
        return battery_level, inspection_log, warning_zones

    # FIX ME: Gantikan [] dengan senarai zones
    for zone in []:
        # FIX ME: Tolak 15% bateri
        battery_level = battery_level - 0

        # FIX ME: Simpan zon sebenar
        inspection_log.append("FIX_ME")

        # FIX ME: Tentukan threshold suhu tinggi
        if zone_readings[zone] > 999:
            warning_zones.append(zone)

        # FIX ME: Hentikan jika bateri kurang daripada 20%
        if battery_level < 0:
            break

    return battery_level, inspection_log, warning_zones
````

# TASK 4: Operasi Utama dan Laporan Akhir

Gunakan senario operasi selamat:

```text
Bateri = 90%
Jarak = 60 cm
```

````python
# ==========================================================
# TASK 4: JALANKAN MISI AKHIR
# ==========================================================

# FIX ME: Tetapkan nilai operasi selamat
operation_battery = None
operation_distance = None

if operation_battery is not None and operation_distance is not None:
    decision = decide_mission(operation_battery, operation_distance)
    final_battery, inspection_log, warning_zones = run_patrol(operation_battery, decision)
else:
    decision = "FIX_ME: TETAPKAN INPUT"
    final_battery, inspection_log, warning_zones = None, [], []

print("\n==============================================")
print("       LAPORAN MISI AKHIR KV-TRON")
print("==============================================")
print("Keputusan       :", decision)
print("Bateri Awal     :", operation_battery)
print("Bateri Akhir    :", final_battery)
print("Zon Diperiksa   :", inspection_log)
print("Zon Suhu Tinggi :", warning_zones)
````

# TASK 5: Ujian Validation Wajib

**Sasaran:** Paparan menunjukkan `4 / 4 UJIAN KESELAMATAN LULUS`.

````python
# ==========================================================
# VALIDATION: JANGAN PADAM CELL INI
# ==========================================================

tests = [
    (90, 60, "MISI DILULUSKAN"),
    (80, 20, "AMARAN: KURANGKAN KELAJUAN"),
    (15, 80, "KEMBALI KE PENGECAS"),
    (75, 5, "HENTI KECEMASAN")
]

passed = 0
for battery, distance, expected in tests:
    actual = decide_mission(battery, distance)
    result = "LULUS" if actual == expected else "BELUM LULUS"
    print(f"Bateri={battery}% | Jarak={distance}cm | Dijangka={expected} | Sebenar={actual} | {result}")
    passed += int(actual == expected)

print("\nHASIL VALIDATION:", passed, "/ 4 UJIAN KESELAMATAN LULUS")

submission_ready = (
    identity_ready
    and passed == 4
    and decision == "MISI DILULUSKAN"
    and len(inspection_log) > 0
)

print("[SEDIA UNTUK DIHANTAR]" if submission_ready else "[BAIKI SEBELUM HANTAR]")
````
