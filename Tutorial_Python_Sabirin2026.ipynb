```python
def gred(markah):

    if markah >= 80:
        return "A"

    elif markah >= 70:
        return "B"

    elif markah >= 60:
        return "C"

    elif markah >= 50:
        return "D"

    else:
        return "F"

markah = int(input("Masukkan markah: "))

print("Gred:", gred(markah))
```

```python
def kira_gred(markah):

    if markah >= 80:
        return "A"
    elif markah >= 60:
        return "B"
    elif markah >= 40:
        return "C"
    else:
        return "Gagal"

bil = int(input("Bilangan pelajar: "))

for i in range(bil):

    nama = input("Nama: ")
    markah = int(input("Markah: "))

    print(nama, "-", kira_gred(markah))
```
